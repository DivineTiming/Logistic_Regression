**Question 1 :**

What is Simple Linear Regression (SLR)? Explain its purpose.

**Answer 1 :**

Simple Linear Regression is a statistical technique that models the relationship between a single independent variable and a dependent variable by fitting a straight line to the observed data.

The main purpose is to predict the value of the dependent variable using the known values of the independent variable and to quantify the strength and direction of their relationship.

 - Predicting outcomes: To estimate the value of the dependent variable for a new value of the independent variable.

 - Quantifying relationships: To measure the strength and direction of the linear association between two variables. For example, one might measure how much job satisfaction increases for every additional dollar in income.

 - Finding a best-fit line: To find the straight line that best represents the data points on a scatter plot, typically by minimizing the sum of the squared differences between the observed and predicted values (the least squares method).

 - Estimating effects: To estimate the value of the dependent variable for a specific value of the independent variable and to understand the change in the dependent variable for a one-unit change in the independent variable.

 - Simplifying complex data: To replace a collection of data points with a single, simple linear equation that captures the main trend.


**Question 2 :**

What are the key assumptions of Simple Linear Regression?

**Answer 2 :**

Assumptions of Simple Linear Regression are :

 - Linearity: The relationship between independent and dependent variables is linear.

 - Independence: Observations are independent of each other.

 - Homoscedasticity: Constant variance of errors across all levels of the independent variable.

 - Normality: Residuals (errors) are normally distributed.


**Question 3 :**

Write the mathematical equation for a simple linear regression model and
explain each term.

**Answer 3 :**

A simple linear regression model describes the relationship between two variables: a dependent variable (Y) and an independent variable (X). The equation is as follows:

$Y = \beta_0 + \beta_1 X + \epsilon$

Representation of the terms:

*   **Y**: This is the **dependent variable** (also known as the response variable or outcome variable). It's the variable we are trying to predict or explain.

*   **X**: This is the **independent variable** (also known as the predictor variable or explanatory variable). It's the variable used to predict Y.

*   **$\beta_0$**: This is the **Y-intercept** of the regression line. It represents the expected value of Y when X is 0. In some contexts, it can be interpreted as the baseline value of Y.

*   **$\beta_1$**: This is the **slope** of the regression line. It represents the change in Y for a one-unit increase in X. It quantifies the strength and direction of the linear relationship between X and Y.

*   **$\epsilon$**: This is the **error term** or residual. It represents the difference between the actual observed value of Y and the value predicted by the model ($ \beta_0 + \beta_1 X $). It accounts for all other factors that affect Y but are not included in the model, as well as random variability. It is assumed to be normally distributed with a mean of zero and constant variance.

**Question 5 :**

What is the method of least squares in linear regression?

**Answer 5 :**

The Method of Least Squares is a statistical procedure to find the best-fitting line for a set of data points by minimizing the sum of the squares of the offsets (residuals) of the points from the line.

The goal is to estimate the coefficients intercept and slope such that the predicted values of Y are as close as possible to the actual observed values of Y.

Here's how it works:

Residuals (Errors): For each data point (xi, yi), the model predicts a value. The difference between the actual observed value and the predicted value is called the residual or error.

Sum of Squared Residuals (SSR): The method of least squares aims to minimize the sum of the squares of these residuals.

Minimization: By using calculus, we can find the values of beta 0 and beta 1 that minimize this SSR. These values are the least squares estimates of the population parameters Beta0 and Beta1.

Why square the residuals?

Squaring ensures that positive and negative errors do not cancel each other out. This means that large errors (positive or negative) are penalized more heavily than small errors.

Squaring makes the function differentiable, which is necessary for calculus-based minimization techniques.

The method of least squares finds the line that passes through the data points in such a way that the vertical distances from each point to the line, when squared and summed, are as small as possible. This line is often referred to as the "line of best fit".


**Question 6 :**

What is Logistic Regression? How does it differ from Linear Regression?

**Answer 6 :**

Logistic Regression is a statistical model used for binary classification problems, meaning it predicts the probability that an observation belongs to one of two categories (e.g., 'yes' or 'no', 'true' or 'false', 'spam' or 'not spam').

It is a classification algorithm, not a regression algorithm in the traditional sense of predicting a continuous value.

The core of logistic regression is the logistic function (also known as the sigmoid function), which squashes any real-valued number into a value between 0 and 1. This output can then be interpreted as a probability.

While both Linear Regression and Logistic Regression are generalized linear models, they serve different purposes and have fundamental differences:

Purpose:

Linear Regression: Used for regression tasks, predicting a continuous numerical output (e.g., house prices, temperature).
Logistic Regression: Used for classification tasks, predicting the probability of a categorical outcome (typically binary, but can be extended to multi-class).

Output:

Linear Regression: Outputs a continuous value, which can range from negative infinity to positive infinity.
Logistic Regression: Outputs a probability value between 0 and 1, which is then mapped to a discrete class label.
Equation/Underlying Function:

Linear Regression: Uses a linear equation.

Logistic Regression: Uses the logistic (sigmoid) function to transform the linear combination of inputs into a probability. It models the log-odds (logit) of the probability of the event occurring as a linear combination of the predictors.

Error Distribution/Assumptions:

Linear Regression:
Assumes that the errors (residuals) are normally distributed and have constant variance (homoscedasticity). The dependent variable is continuous.
Logistic Regression:
Assumes that the dependent variable follows a Bernoulli distribution (for binary outcomes). It does not assume normality or homoscedasticity of residuals.
Cost Function/Optimization:

Linear Regression:
Typically uses Mean Squared Error (MSE) as its cost function, minimized through methods like Ordinary Least Squares (OLS).
Logistic Regression:
Typically uses Maximum Likelihood Estimation (MLE) to find the optimal coefficients, often involving cost functions like cross-entropy loss, which are designed to penalize incorrect probability predictions more heavily.


** Question 7 :**

Name and briefly describe three common evaluation metrics for regression models.

** Answer 7 :**

Three common evaluation metrics for regression models are :

 - **Mean Absolute Error (MAE)**:

 MAE measures the average magnitude of the errors in a set of predictions, without considering their direction. It is the average of the absolute differences between the predicted and actual values. MAE is a linear score, which means all individual differences are weighted equally in the average.

 A lower MAE indicates a better fit of the model to the data. It's robust to outliers because it doesn't square the errors.

 - **Mean Squared Error (MSE)**:

MSE measures the average of the squares of the errors. It is the average squared difference between the predicted and actual values. By squaring the errors, MSE penalizes larger errors more heavily than smaller ones.

Like MAE, a lower MSE indicates a better model. Due to the squaring of errors, it is more sensitive to outliers than MAE. The units of MSE are the square of the units of the target variable.

 - **R-squared (Coefficient of Determination)**:

R-squared is a statistical measure that represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s). It indicates how well the model explains the variability of the response variable around its mean.

R-squared values range from 0 to 1 (or sometimes negative if the model is worse than a horizontal line). An R-squared of 1 indicates that the model perfectly explains the variance of the target variable, while an R-squared of 0 indicates that the model explains no variance. A higher R-squared generally indicates a better fit, but it can be misleading if more independent variables are added, even if they don't improve the model's predictive power (for this, Adjusted R-squared is often preferred).


** Question 8 :**

What is the purpose of the R-squared metric in regression analysis?

** Answer 8 :**

The R-squared (R^2) metric, also known as the Coefficient of Determination, is a key evaluation metric in regression analysis that serves the following purpose:

Explains Variance: Its primary purpose is to represent the proportion of the variance in the dependent variable (the variable your model is trying to predict) that can be explained by the independent variable(s) (the predictor variables) in your regression model.

Goodness of Fit: It provides a measure of how well the regression model fits the observed data. A higher R-squared value indicates that the model accounts for a larger proportion of the variance, implying a better fit to the data.

An R-squared of 1 (or 100%) means the model perfectly explains all the variability in the dependent variable. This is rare in real-world scenarios.
An R-squared of 0 means the model explains none of the variability in the dependent variable around its mean. In essence, the model is no better than simply predicting the mean of the dependent variable for every observation.
Values between 0 and 1 indicate the percentage of variance explained.
Model Performance Indicator: It helps in understanding the predictive power of the model. If a model has an R-squared of, say, 0.75, it means that 75% of the variation in the dependent variable can be predicted from the independent variables.

Considerations:

Correlation vs. Causation: R-squared does not imply causation. A high R-squared only indicates a strong statistical relationship, not necessarily a causal one.
Overfitting: Adding more independent variables to a model will almost always increase R-squared, even if the new variables are not truly related to the dependent variable. This can lead to overfitting. For this reason, Adjusted R-squared is often preferred, as it accounts for the number of predictors in the model and penalizes the addition of unnecessary variables.
Context Matters: What constitutes a 'good' R-squared value varies significantly by field. In some experimental sciences, an R-squared of 0.9 may be expected, while in social sciences, an R-squared of 0.3 could be considered quite good.


** Question 10 :**

 How do you interpret the coefficients in a simple linear regression model?

 ** Answer 10 :**

In a simple linear regression model, the equation is typically represented as:

$Y = \beta0 + \beta1 X + \epsilon$

Where:
**Y** is the dependent variable (the outcome you're trying to predict).
**X** is the independent variable (the predictor).
**$\beta0$** is the intercept.
**$\beta1$** is the slope (the coefficient for X).
**$\epsilon$** is the error term.


 - Intercept ($\beta0$)

**Interpretation**: The intercept represents the **expected mean value of the dependent variable (Y) when the independent variable (X) is equal to 0.**

**Contextual Relevance**: Its interpretation depends heavily on whether an X value of 0 is meaningful or plausible in the context of your data.

**Meaningful Zero**: If X can realistically be 0 (e.g., hours studied), then the intercept is directly interpretable. For example, if X is 'hours studied' and Y is 'exam score', an intercept of 60 means that a student who studies 0 hours is predicted to score 60.

**Not Meaningful Zero**: If X cannot realistically be 0 (e.g., height, temperature in Celsius where 0 is arbitrary), or if X=0 is outside the range of your observed data, then the intercept might not have a practical or meaningful interpretation on its own. In such cases, it primarily serves to adjust the regression line to the correct level.

 - Slope ($\beta1$)

**Interpretation**: The slope ($\beta1$) represents the **average change in the dependent variable (Y) for every one-unit increase in the independent variable (X)**, assuming all other factors are held constant (though in simple linear regression, there are no other factors).

**Direction**: The sign of the slope indicates the direction of the relationship:

**Positive Slope ($ \beta_1 > 0 $)**: As X increases, Y tends to increase. This indicates a positive linear relationship.

**Negative Slope ($ \beta_1 < 0 $)**: As X increases, Y tends to decrease. This indicates a negative linear relationship.

**Zero Slope ($ \beta_1 = 0 $)**: There is no linear relationship between X and Y.

**Magnitude**: The absolute value of the slope indicates the strength of the relationship (how much Y changes per unit change in X). A larger absolute value means a steeper slope and a stronger effect.


# Practical Question

```python
# Question 9 : Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.


import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Creating a dataset where y = 2*x + 1 + noise
X = np.random.rand(100, 1) * 10 # 100 samples, 1 feature, values from 0 to 10
y = 2 * X + 1 + np.random.randn(100, 1) * 2 # y = 2x + 1 + some noise

# Creating a Linear Regression model
model = LinearRegression()

# Fitting the model to the data
model.fit(X, y)

# Printing the slope (coefficient) and intercept
print(f"Slope (Coefficient): {model.coef_[0][0]:.4f}")
print(f"Intercept: {model.intercept_[0]:.4f}")

# Visualizing the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='green', label='Actual Data')
plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('X (Independent Variable)')
plt.ylabel('y (Dependent Variable)')
plt.title('Simple Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```
